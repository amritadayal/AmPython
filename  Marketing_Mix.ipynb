```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

**Importing data from excel sheel**

```python
df = pd.read_csv("marketing_data.csv")
print(df.head())
```

**Data Cleaning**

```python
df.iloc[:,4] = df.iloc[:,4].str.replace('$', '').str.replace(',', '').str.replace('.00', '')
fill_value = '0'
df.iloc[:,4].fillna(fill_value,inplace=True)
```

Remove "$" sign from the Income Coloum and fill the empty space with value 0

```python
current_year = datetime.now().year
df['Age'] = current_year - df['Year_Birth']

# Display the DataFrame
print(df)
```

Create a new coloum "Age" correspoding to coloum "Year of Birth" of Customer

```python
df.columns = df.columns.str.strip()
income_column = df['Income']
print(income_column)
```

```python
# Calculate the sum of each column
total_values = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum()

# Plot a pie chart for the total values
total_values.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Total Values')
plt.ylabel('')  # Remove the y-label
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

**From above pie chat we can say maximum purchaes is Wine and minimum purchases product is fruits**

As per above pie char representation of the total purchases Wine products are performing the best, and fruits are performing the least.
